**Mpg Case-Study**

```
import io
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
from google.colab import files
uploaded = files.upload()
```

```
mpg=pd.read_csv(io.BytesIO(uploaded['mpg.csv']))
```

```
mpg.head()
```

```
sns.relplot(x='acceleration' ,y='mpg', data=mpg, kind ='scatter' ,style="origin",hue="origin")
plt.show()
```

```
sns.relplot(x='model_year' ,y='mpg', data=mpg, kind ='scatter' )
plt.show()
```

```
sns.relplot(x='model_year' ,y='mpg', data=mpg, kind ='line',ci=None,style="origin",hue="origin",markers=True )
plt.show()
```
